```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class SimpleMLP(nn.Module):
    def __init__(self):
        super(SimpleMLP, self).__init__()
        self.fc1 = nn.Linear(10,20)
        self.fc2 = nn.Linear(20,10)
    def forward(self, x):
        x = F.relu(self.fc1(x))
        return self.fc2(x)
 
```

```python
model = SimpleMLP()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
X = torch.randn(100,10)
y = torch.rand(100,1)
X,y
```

```python
for epoch in range(20):
    output = model(X)
    loss = criterion(output, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

Training done now saving the model

```python
torch.save(model.state_dict(), 'data/simple_mlp.pth')
```

```python
loadedModel = SimpleMLP()
loadedModel.load_state_dict(torch.load('data/simple_mlp.pth'))
```

```python
loadedModel.eval()
# Now do your inference
```
